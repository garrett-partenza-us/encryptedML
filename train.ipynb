```python
import pandas as pd
```

```python
dataset = pd.read_csv('insurance.csv')

dataset.head()
```

```python
dataset.sex = dataset.sex.map({'female': 1, 'male': 0})

dataset.head()
```

```python
dataset.bmi = dataset.bmi.round().astype(int)

dataset.head()
```

```python
dataset.smoker = dataset.smoker.map({'yes': 1, 'no': 0})

dataset.head()
```

```python
dataset.region = dataset.region.map({'southwest': 0, 'southeast': 1, 'northwest': 2, 'northeast': 3})

dataset.head()
```

```python
dataset.charges = dataset.charges.round().astype(int)

dataset.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
y = dataset.charges
X = dataset.drop('charges', axis=1)

model = LinearRegression()
model.fit(X, y)

model.score(X, y)
```

```python
model.coef_
```

```python
model.predict([[19, 1, 27, 0, 0, 1]])
```

```python
import csv
```

```python
# Save the coefficients and feature names to a CSV file
with open('linear_regression_model.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Feature Name', 'Coefficient'])  # Write header row
    for feature, coefficient in zip(X.columns, model.coef_):
        writer.writerow([feature, int(coefficient)])
```
